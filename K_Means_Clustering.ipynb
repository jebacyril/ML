### 7.3 Programming Task: Color Quantization using K-Means clustering.

 The goal of this task is to perform color quantization using the K-Means clustering algorithm.

   i\. Implement the K-Means clustering algorithm using Numpy/SciPy.




```python
def K_means_clustering(train_X):
   
   #TODO: Implement the K means algorithm. The input train_X is an array of shape (p,3), where p is the number of pixels in the image.
   model = KMeans(n_clusters=64,n_init=10)
   target = model.fit_predict(train_X)
   cluster_centers = model.cluster_centers_ 

   return target, cluster_centers

```

   ii\. Read the given image. Consider each pixel as a 3-dimensional vector and run your K-Means clustering algorithm to get 64 clusters.


```python
#TODO
original_image = cv2.imread("sup_7_3_input.jpg")
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title("Given Image")
plt.axis("off")
reshaped_image = np.reshape(original_image,((original_image.shape[0] * original_image.shape[1]), 3))

target, cluster_centers = K_means_clustering(reshaped_image)
print(cluster_centers.shape)

```

  iii\. Assign each pixel the color value of its nearest cluster center. Visualize the result.

```python
#TODO
output_image = np.reshape(cluster_centers.astype("uint8")[target], (original_image.shape[0], original_image.shape[1], 3))
image = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)
plt.title("Color Quantization using K means Clustering")
plt.imshow(image)
plt.axis("off")
```
